The goal of this assignment is to build a simple, modular, extensible, machine learning
pipeline in Python. The task here is to predict who will experience financial distress in the next two years. The outcome variable (label) in the data is SeriousDlqin2yrs.

```python
import create_pipeline as cp
import pandas as pd

pd.options.display.float_format = '{:,}'.format
```

```python
# Specify dataset-specific parameters
csv_filename = 'credit-data.csv'
index_col = 'PersonID'
col_types = {'SeriousDlqin2yrs': int, 
             'RevolvingUtilizationOfUnsecuredLines': float, 
             'age': int, 
             'zipcode': int, 
             'NumberOfTime30-59DaysPastDueNotWorse': int,  
             'DebtRatio': float, 
             'MonthlyIncome': int, 
             'NumberOfOpenCreditLinesAndLoans': int, 
             'NumberOfTimes90DaysLate': int, 
             'NumberRealEstateLoansOrLines': int, 
             'NumberOfTime60-89DaysPastDueNotWorse': int, 
             'NumberOfDependents': int} 
```

### Step 1: Read Data
This step loads a CSV file.

```python
# Load data 
df = cp.load_data(csv_filename)
```

### Step 2: Explore Data
This step generates plots showing distributions of the variables by the target (i.e., SeriousDlqin2yrs). These plots can be useful in suggesting potential relationships between each variable and the target, along with in identifying potential outliers and data anamolies. This step also creates a summary table showing the most correlated variables.

```python
# Specify continuous and categorical variables 
continuous_cols = ['age', 
                   'MonthlyIncome', 
                   'RevolvingUtilizationOfUnsecuredLines', 
                   'DebtRatio']
categorical_cols = ['zipcode', 
                    'NumberOfDependents', 
                    'NumberOfTime30-59DaysPastDueNotWorse', 
                    'NumberOfTime60-89DaysPastDueNotWorse', 
                    'NumberOfTimes90DaysLate', 
                    'NumberOfOpenCreditLinesAndLoans', 
                    'NumberRealEstateLoansOrLines']
target_col = 'SeriousDlqin2yrs'
```

```python
# Plot continuous variables 
for v in continuous_cols: 
    df.boxplot(column=[v], by=target_col, figsize=(20, 5), vert=False)
```

```python
# Plot categorical variables
for v in categorical_cols: 
    table = pd.crosstab(df[v], df[target_col])
    table.div(table.sum(1).astype(float), axis=0).plot(kind='bar', stacked=True, figsize=(20, 5))
```

```python
# Generate correlations between variables 
df.corr().abs().unstack().sort_values(ascending=False).round(3)[:25]
```

### Step 3: Process Data
This step sets does a few data processing steps including setting the dataframe's index, replacing missing values with the column's median value, and setting the data types for the columns (as specified in the dataset parameters above). Missing values can also be set to the column's mean, or they can be dropped altogether (by not specifying this optional parameter). 

```python
df = cp.process_data(df, index_col, 'median', col_types)
```

### Step 4: Generate Features
This step creates categorical and binary variables from the existing variables in the data. Specifically, it creates the following categories: age_cat, MonthlyIncome_cat, RevolvingUtilization_cat, DebtRatio_cat, and OpenCredit_cat. It splits the data into four bins, but this parameter can be specified. 

This step also creates the following binary variables: EverPastDue_binary, HasDependents_binary, RealEstateLoans_binary. These are specified by a cutoff value, which is also a parameter that can be specified. 

```python
# Convert continuous variables to categorical variables 
df = cp.continuous_to_categorical(df, 'age_cat', 'age', 4)
df = cp.continuous_to_categorical(df, 'MonthlyIncome_cat', 'MonthlyIncome', 4)
df = cp.continuous_to_categorical(df, 'RevolvingUtilization_cat', 'RevolvingUtilizationOfUnsecuredLines', 4)
df = cp.continuous_to_categorical(df, 'DebtRatio_cat', 'DebtRatio', 4)
df = cp.continuous_to_categorical(df, 'OpenCredit_cat','NumberOfOpenCreditLinesAndLoans', 4)

# Create new variable for total number of times past due 
df['TotalNumberOfTimesPastDue'] = df['NumberOfTime30-59DaysPastDueNotWorse'] + df['NumberOfTime60-89DaysPastDueNotWorse'] + df['NumberOfTimes90DaysLate'] 

# Convert categorical variables to binary variables 
df = cp.categorical_to_binary(df, 'EverPastDue_binary', 'TotalNumberOfTimesPastDue', 0)
df = cp.categorical_to_binary(df, 'HasDependents_binary', 'NumberOfDependents', 0)
df = cp.categorical_to_binary(df, 'RealEstateLoans_binary', 'NumberRealEstateLoansOrLines', 1)
```

### Step 5: Build Classifier
This step first splits the data into a training and testing dataset, and then creates both a logistic regression and decision tree classifier. In addition to specifying the dataframe, target column, and attribute column, this also allows for parameters to specify the proportion of data to be included in the testing and training dataset, along with a random seed. 

Both classifiers also take a random seed parameter. The logistic regression classifier also allows for specification of the solver (or optimizing algorithm). The decision tree classifier allows for specification of the split quality function (gini, entropy), the tree's maximnum depth, and the minimum number of observations required for a split. 

```python
# Build classifiers 
feature_cols = ['zipcode', 'age_cat', 'MonthlyIncome_cat', 'RevolvingUtilization_cat', 
                'DebtRatio_cat', 'OpenCredit_cat', 'EverPastDue_binary', 
                'HasDependents_binary', 'RealEstateLoans_binary']

x_train, x_test, y_train, y_test = cp.split_data(df, feature_cols, target_col, 0.2, 123)
logreg_classifier = cp.build_logreg_classifier(123, 'liblinear')
dt_classifier = cp.build_tree_classifier(123, 'entropy', 10, 10)
```

### Step 6: Evaluate Classifiers
This step evaluates the two classifiers built above based on their accuracy in correctly predicting the target variable in the testing dataset. Using the model specified, the logistic regression classifier had an accuracy score of around 83%, while the decision tree classifier had an accuracy score of around 85%.  

```python
# Evaluate classifiers
print('Logistic Regression:')
cp.evaluate_classifier(logreg_classifier, x_train, y_train, x_test, y_test)
print('Decision Tree:')
cp.evaluate_classifier(dt_classifier, x_train, y_train, x_test, y_test)
```
